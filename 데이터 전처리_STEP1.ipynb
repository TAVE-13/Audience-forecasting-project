# I. 애니메이션을 제외한 데이터 전처리

```python
data = pd.read_excel("boxoffice.xlsx")
data
```

```python
data.info()
```

```python
# 데이터 결측치 수 확인
data_missing_values = data.isnull().sum()

print(f'data_missing values: \n{data_missing_values}')
```

```python
data.drop(columns=['순번', '인스타', '블로그', 'X(트위터)'], inplace=True)
```

```python
columns_to_fill = ['배우 언급수', '감독 평점', '평론가 평점', '총 언급수', '조회수']
data[columns_to_fill] = data[columns_to_fill].fillna(0)
```

```python
columns_to_convert = ['배우 언급수','총 언급수','조회수']
data[columns_to_convert] = data[columns_to_convert].astype(int)
```

```python
data
```

```python
output_file_path = "boxoffice_data.xlsx"
data.to_excel(output_file_path, index=False)
```

## I-A. 범주형 변수 정수 인코딩

```python
data['상영등급_encoded'], rating_categories = pd.factorize(data['상영등급'])
data['장르_encoded'], genre_categories = pd.factorize(data['장르'])

rating_mapping_df = pd.DataFrame({
    '상영등급': rating_categories,
    '상영등급_encoded': range(len(rating_categories))
})

genre_mapping_df = pd.DataFrame({
    '장르': genre_categories,
    '장르_encoded': range(len(genre_categories))
})

data.drop(columns=['상영등급', '장르'], inplace=True)

data.rename(columns={'상영등급_encoded': '상영등급', '장르_encoded': '장르'}, inplace=True)
```

```python
data
```

```python
rating_mapping_df = pd.DataFrame({
    '상영등급': rating_categories,
    '상영등급_encoded': range(len(rating_categories))
})

genre_mapping_df = pd.DataFrame({
    '장르': genre_categories,
    '장르_encoded': range(len(genre_categories))
})

print("상영등급 매핑:")
print(rating_mapping_df)

print("\n장르 매핑:")
print(genre_mapping_df)
```

```python
output_file_path = "data_encoded.xlsx"
data.to_excel(output_file_path, index=False)
```
